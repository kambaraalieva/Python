```python
mark = float(input("Введите оценку: "))

if mark >= 90:
    print("Отлично, Ваша оценка 5!")
elif 80 <= mark < 90:
    print("Здорово, Ваша оценка 4!")
elif 70 <= mark < 80:
    print("Хорошо, Ваша оценка 3!")
elif 60 <= mark < 70:
    print("Вам стоит подучить материал")
else:
    print("Вы не сдали экзамен")
```

```python
x = int(input("x = "))
y = int(input("y = "))
z = int(input("z = "))

if x <= y and x <= z:
    print(x)
elif y <= x and y <= z:
    print(y)
else:
    print(z)
```
