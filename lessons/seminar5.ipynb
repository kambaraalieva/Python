```python
names = ['Александр', 'Наполеон', 'Лао']
surnames = ['Македонский','Бонапарт','Цзы']

for n, s, i in zip(names, surnames, range(len(names))):
    names[i] = n + ' ' + s

print(names)
```

```python
#task19
counter = 1
counter_global = 1

DNA = 'AATTAGGATGATCGCTAGGCTCGATAGGCTAGTTCCGGGGGGGAGCTGACTGCCCCCCCCCC'
for i in range(len(DNA) - 1):
    if DNA[i] == DNA[i + 1]:
        counter += 1
        counter_global = max(counter_global, counter)
    else:
        counter = 1

print(counter_global)
```

```python
list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]
result = []
sum_ = []

for i in range(len(list1)):
    sum_.append(list1[i] + list2[i])
    result.append(list1[i] + list2[-(i + 1)])

print(sum_)
print(result)
```
