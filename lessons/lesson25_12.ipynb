```python
def log_events(**data):
    for key, value in data.items():
        print(f"{key}: {value}")
```

```python
log_events(user= "Zhamilya", action= "login", status= "ok")
```

```python
def demo(*args, **kwargs):
    print(f"Args {args}")
    print(f"Kwargs {kwargs}")
```

```python
demo(1, 2, 3, name="Aiana", age=89)
```

```python
def count_down(n):
    if n == 0:
        return
    print(n)
    count_down(n-1)

count_down(17)
```

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

```python
def read_file(path):
    with open(path, "r", encoding= "utf-8") as file:
        for line in file:
            print(line.strip())
```

```python
def sum_numbers(*args):
    total = 0
    for number in args:
        total += number
    return total
```

```python
print(sum_numbers(3, 4, 6, 7, 8, 9))
```

```python
class Employee:
    def __init__(self, salary):
        self._salary = salary

    def get_salary(self):
        return self._salary

    def get_role(self):
        return "Employee"


class Manager(Employee):
    def get_salary(self):
        return self._salary + 1000

    def get_role(self):
        return "Manager"
```

```python
class Developer(Employee):
    def get_salary(self):
        return self._salary + 1000000

    def get_role(self):
        return "Developer"
```

```python
employees = {
    Employee(1000),
    Developer(2000),
    Manager(2000),
}
```

```python
for employee in employees:
    print(f"Role: {employee.get_role()}, Salary: {employee.get_salary()}")
```
