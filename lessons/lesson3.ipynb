```python
def two_natural(a):
    count = 0
    for num in range(2, a):
        if a % num == 0:
            count+=1
    if count == 2:
        return True
    return False
```

```python
list(filter(two_natural, range(1, 10)))
```

```python
def two_natural_from_range(a, b):
    list_ = list(range(a, b+1))

    return list(filter(two_natural, list_))
```

```python
two_natural_from_range(174457, 174505)
```

```python
def four_natural(a):
    count = 0
    for num in range(2, a):
        if a % num == 0:
            count+=1
    if count == 4:
        return True
    return False
```

```python
def four_natural_from_range(a, b):
    list_ = list(range(a, b+1))
    return list(filter(four_natural, list_))
```

```python
def four_natural_dict(a, b):
    list_ = four_natural_from_range(a, b)
    dict_={}
    for n in list_:
        temp = []
        for i in range(2, n):
            if n % i == 0:
                temp.append(i)
        dict_[n] = temp
    return dict_
```

```python
four_natural_dict(2, 100)
```
