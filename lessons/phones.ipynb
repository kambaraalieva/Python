```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.sulpak.kg/f/noutbuki/bishkek/846_48'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.find('div', class_="product__item-price")
```

```python
len(soup.find_all('div', class_="product__item product__item-js tile-container"))
```

```python
laptops = soup.find('div', id='products').find_all('div', class_="product__item product__item-js tile-container")
```

```python
laptops[0].find_all('a')
```

```python
for laptop in laptops:
  name = laptop['data-name'].split(' / ')[0].replace('Ноутбук', '')
  price = laptop['data-price']
  url = laptop.find('div', class_='product__item-name').a['href']
  print(name, price, url)
```
