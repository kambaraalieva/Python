```python
dict_ = {
    'n_1': [23, 31, 45, 67, 12, 24],
    'n_2': [23, 31, 45, 67, 10, 24],
    'n_3': [23, 31, 45, 67, 11, 24],
}
```

```python
def dict_avg(dict_):
    key, value = dict_
    return key, round(sum(value) / len(value))
```

```python
dict(map(dict_avg, dict_.items()))
```

```python
def is_float_in(dict_):
    key, value = dict_
    return key, isinstance(sum(value), float)
```

```python
dict(map(is_float_in, dict_.items()))
```

```python
def more_floats_than_int(dict_2):
    count_float = 0
    count_int = 0
    k, v = dict_2
    for value in v:
        if type(value) == float:
            count_float+=1
        else:
            count_int+=1
    if count_int < count_float:
        return k, True
    else:
        return k, False
```

```python
dict_2 = {
    'n_1': [23, 3.1, 45, 6.7, 1.2, 6.5],
    'n_2': [23, 31, 45, 24, 5.6, 6.7, 8.9, 7.0, 3.4 ],
    'n_3': [23, 31, 45, 67, 11, 24],
}
```

```python
dict(map(more_floats_than_int, dict_2.items()))
```

```python
numbers = {
    'n_1': [23, 3, 45, 6, 1, 6],
    'n_2': [23, 31, 45, 8, 7, 3],
    'n_3': [23, 31, 45, 67, 11, 24],
}
```

```python
def med(n):
    k, v  = n
    mid = len(v) // 2
    v.sort()
    if (len(v) % 2 == 1):
        return k,  v[len(v)//2]
    elif len(v) % 2 == 0:
        return k, (v[mid] + v[mid - 1]) // 2
```

```python
median = dict(map(med, numbers.items()))
```

```python
print(median)
```

```python
products = [
    {"name": "apple", "price": 100},
    {"name": "banana", "price": 40},
    {"name": "mango", "price": 120},
]
```

```python
updated = map(lambda x: {"name": x["name"], "price": x["price"] * 1.15}, products)
```

```python
filtered = list(filter(lambda x: x["price"] > 100, updated))
```

```python
final = list(map(lambda x: x["name"], filtered))
print(final)
```
