```python
def is_just(n):
    i = 0
    for i in range(2, n + 1):
        if n % i == 0:
            i = n + 1
    if i > 1:
        return "It is not just"
    elif n == 1:
        return "It is neither just nor unjust"
    else:
        return "JUst"
```

```python
def get_matrix(n):
    list_ = [[i for i in range(1, n+1)] for i in range (n)]
    if n == 0:
        return []
    if n < 0:
        return [[i for i in range(abs(n), 0, -1)] for i in range (abs(n))]
    if n > 0:
        return list_
```

```python
get_matrix(0)
```

```python
def range_matrix(n):
    if (n ** (1/2) % 1 == 0):
        k = int(n ** (1/2))
        list_ = []
        for i in range(1, n, k):
            temp = []
            for j in range(i, i+k):
                temp.append(j)
            list_.append(temp)
        return list_
    else:
        return []
```

```python
range_matrix(9)
```

```python
def range_matrix_snake(n):
    if (n ** 0.5 % 1 == 0):
        k = int(n ** 0.5)
        matrix = []
        for i in range(0, n, k):
            row = list(range(i + 1, i + k + 1))
            if len(matrix) % 2 == 1:
                row.reverse()
            matrix.append(row)
        return matrix
    else:
        return []
```

```python
range_matrix_snake(9)
```
