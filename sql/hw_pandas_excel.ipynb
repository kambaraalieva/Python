```python
import pandas as pd

# Читаем файл sales.xlsx
df = pd.read_excel("sales.xlsx")

# Выводим таблицу на экран
print(df)
```

```python
# Создаем новый столбец total_price
df['total_price'] = df['price'] * df['quantity']

# Проверим, что столбец добавился
print(df.head())
```

```python
# Общая сумма продаж
total_sales = df['total_price'].sum()
print("Общая сумма продаж:", total_sales)

# Средняя цена товара
average_price = df['price'].mean()
print("Средняя цена:", average_price)
```

```python
electronics_df = df[df['category'] == 'Electronics']

# Выводим только электронику
print(electronics_df)
```

```python
# Сохраняем весь DataFrame с новым столбцом в Excel
df.to_excel("sales_result.xlsx", index=False)

# Если нужно сохранить только Electronics:
electronics_df.to_excel("electronics_result.xlsx", index=False)
```
