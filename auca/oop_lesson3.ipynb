```python
class CandyBox:
    def __init__(self, candies):
        self.candies = candies

    def get_candies(self):
        return self._candies

    def set_candies(self, value):
        if value < 0:
            print("No, candies cannot be negative")
            return
        self._candies = value

a = CandyBox(5)
a. set_candies(5)
a.set_candies(-9)
```

```python
class CandyBox:
    def __init__(self, candies):
        self._candies = 0
        self.candies = candies

    @property
    def candies(self):
        return self._candies

    @candies.setter
    def candies(self, value):
        if value < 0:
            print("No, candies cannot be negative")
            return
        self._candies = value
```

```python
class Box:
    def __init__(self):
        self._weight = 10

    @property
    def weight(self):
        return self._weight
```

```python
b = Box()
print(b.weight)
```

```python
class Animal:
    def speak(self):
        print("Животное издает звук")

class Dog(Animal):
    def speak(self):
        print("Гав")

class Cat(Animal):
    def speak(self):
        print("Мяу")

animals = [Dog(), Cat(), Animal()]

for a in animals:
    a.speak()
```
