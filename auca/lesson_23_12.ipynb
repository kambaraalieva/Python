```python
class BankAccounts:
    def __init__(self, accounts):
        self.accounts = accounts
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.accounts):
            raise StopIteration

        account = self.accounts[self.index]
        self.index += 1
        return account
```

```python
data = [
    {"client": "Aijan", "balance": 50000},
    {"client": "Bek", "balance": 5},
    {"client": "Aia", "balance": 50000},
]
```

```python
accounts = BankAccounts(data)
```

```python
for acc in accounts:
    print(acc["client"])
```
