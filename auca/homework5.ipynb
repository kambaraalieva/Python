```python
# task19
best = 0
best_girls = 0

for g1 in range(26):
    g2 = 25 - g1
    b1 = 22 - g1
    b2 = 20 - b1

    if b1 >= 0 and b2 >= 0:
        total = (g1 / 22) * 100 + (g2 / 23) * 100
        if total > best:
            best = total
            best_girls = g1

print(best_girls, 25 - best_girls)
print(best)
```
