```python
def mult(list1):
    res = 1
    for i in list1:
        res = res * i
    return res
```

```python
list2 = [1, 2, 3, 7]
mult(list2)
```

```python
print(mult([1, 2, 3, 4, 5]))
```

```python
users = [
  { 'name': 'Ron', 'age': 18, 'work': 'IT-backend developer'},
  { 'name': 'Sam', 'age': 23, 'work': 'Doctor'},
  { 'name': 'Bob', 'age': 26, 'work': 'Driver'},
  { 'name': 'John', 'age': 39, 'work': 'Teacher'},
  { 'name': 'Jack', 'age': 29, 'work': 'IT-frontend developer'}
]
```

```python
def employee(users):
    result = []
    for user in users:
        if user["age"] > 21 and user["work"] in ("IT-backend developer", "IT-frontend developer"):
            result.append(user)
    return result
```

```python
employee(users)
```

```python
def check_nan(list_):
    result = []
    for i in list_:
        if i != 'nan':
            result.append(i)
    return result
```

```python
list1 = [5, 6, 7, 8, 9, 2, 3, 'nan', 7, 8, 'nan']
check_nan(list1)
```
