```python
import requests
from bs4 import BeautifulSoup

URL = "https://intermedia.kg/catalog/noutbuki_i_aksessuary/noutbuki_i_ultrabuki/"
```

```python
response = requests.get(URL)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
category = "Ноутбуки и ультрабуки"
```

```python
laptops = []
```

```python
blocks = soup.find("div", class_="items productList")
```

```python
if blocks:
    items = blocks.find_all("div", class_="productColText")

    for item in items:
        a_tag = item.find("a", class_="name")
        name = a_tag.get_text(strip=True)
        link = "https://intermedia.kg" + a_tag.get("href")

        price_tag = item.find("a", class_="price")
        if price_tag:
            # Get only direct text, strip, remove spaces and 'c'
            price_text = ''.join(price_tag.find_all(string=True, recursive=False)).strip()
            price_text = price_text.replace("\xa0", "").replace("c", "")

            # Keep only digits at the start (ignore / шт or extra text)
            cleaned_price = ""
            for char in price_text:
                if char.isdigit():
                    cleaned_price += char
                else:
                    break
            try:
                price = float(cleaned_price)
            except ValueError:
                price = 0
        else:
            price = 0

        laptops.append({
            "category": category,
            "name": name,
            "price": price,
            "url": link
        })
```

```python
for i, laptop in enumerate(laptops, start=1):
    print(f"{i}. {laptop}")
```
