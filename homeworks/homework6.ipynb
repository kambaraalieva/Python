```python
z1_options = [(3, 0), (0, 1)]
z2 = (1, 1)

max_products = -1
best_combination = None

for a1, b1 in z1_options:
    a2, b2 = z2
    total_A = a1 + a2
    total_B = b1 + b2

    products = min(total_A, total_B // 3)

    if products >= max_products:
        max_products = products
        best_combination = ((a1, b1), (a2, b2), (total_A, total_B))

if best_combination is not None:
    print("Максимальное число изделий:", max_products)
    print("Завод 1:", best_combination[0])
    print("Завод 2:", best_combination[1])
    print("Итого:", best_combination[2])
```
