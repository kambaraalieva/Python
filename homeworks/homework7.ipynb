```python
dict4 = {'a': 1, 'b': 4, 'c': 1, 'd': 5, 'e': 6}
```

```python
for key, value in dict4.copy().items():
    if value < 4:
        del dict4[key]
```

```python
dict4 #task4
```

```python
dict5 = {1: {'a': 1}, 2: {'b': 4}, 3: {'c': 1}, 4: {'d': 5}, 5: {'e': 6}}
```

```python
for key, value in dict5.copy().items():
    dict5[key] = list(value.keys())[0]
```

```python
dict5  #task5
```
