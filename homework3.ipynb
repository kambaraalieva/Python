```python
number = (input('Введите номер региона: '))

number = int(number)
if number == 0o1:
    print('Бишкек\n')
elif number == 0o2:
    print('Ош\n')
elif number == 0o3:
    print('Баткенская область\n')
elif number == 0o4:
    print('Джалал-Абадская область\n')
elif number == 0o5:
    print('Нарынская область\n')
elif number == 0o6:
    print('Таласская область\n')
elif number == 0o7:
    print('Чуйская область\n')
else:
    print('Номер неправильно введен ')
```

```python
string7 = input('Enter a number, a text or other: ')  #task7

try:
    number = int(string7)
    print(f'{number} is an integer\n')
except ValueError:
    try:
        number = float(string7)
        print(f'{number} is an float\n')
    except ValueError:
        print(f'{number} is a text\n')
```
